```python
# importing seaborn (for advanced visualisation)
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize']= 8,8
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
visual1 = sns.distplot(stats['InternetUsers'], bins = 40)      #distance plot
```

```python
visualbox1 = sns.boxplot(data = stats, x ="IncomeGroup", y= "BirthRate")  #boxplot
plt.rcParams['figure.figsize']=10,2
```

```python
visual3 = sns.lmplot(data=stats, x='InternetUsers', y= 'BirthRate')   #linear model(lm) plot
```

```python
visual4 = sns.lmplot(data=stats, x='InternetUsers', y= 'BirthRate', fit_reg= False)  #lm without the regression line
```

```python
visual5 = sns.lmplot(data=stats, x='InternetUsers', y= 'BirthRate', fit_reg= False, hue ='IncomeGroup')   #lm plot with colour code according to income group
```
